# PyCity School Analysis

```python
import pandas as pd
```

```python
#files to load
schoolDataToLoad = ("Resources/schools_complete.csv")
studentDataToLoad = ("Resources/students_complete.csv")
```

```python
#read files and store into pandas df's
schoolData = pd.read_csv(schoolDataToLoad)
studentData = pd.read_csv(studentDataToLoad)
```

```python
#merge the 2 data data sets into 1
schoolDataMerged = pd.merge(schoolData, studentData, how="left", on=["school_name", "school_name"])
schoolDataMerged.head()
```

# District Summary

```python
#total Schools in district
schoolCount = len(schoolDataMerged["school_name"].unique())
#total students in district
studentCount = schoolDataMerged ["student_name"].count()
studentCount
```

```python
#total budget for district
totalBudget = schoolDataMerged["budget"].sum()
totalBudget
```

```python
#average math score
averageMathScore = schoolDataMerged["math_score"].mean()
#average reading score
averageReadingScore = schoolDataMerged["reading_score"].mean()
```

```python
#percent passing math
studentsPassingMath = schoolDataMerged[(schoolDataMerged["math_score"] >= 70)].count()["student_name"]
percentPassingMath = studentsPassingMath/ float(studentCount)*100
percentPassingMath
```

```python
#percent passing reading
studentsPassingReading = schoolDataMerged[(schoolDataMerged["reading_score"] >= 70)].count()["student_name"]
percentPassingReading = studentsPassingReading/ float(studentCount)*100
percentPassingReading
```

```python
#percent passing math and reading
studentsPassingBoth = schoolDataMerged[(schoolDataMerged["math_score"] >= 70 ) & (schoolDataMerged["reading_score"] >= 70 )] .count()["student_name"]
percentPassingBoth = studentsPassingBoth/ float(studentCount)*100
percentPassingBoth
```

```python
districtSummary = pd.DataFrame({
    "Total Schools" : [schoolCount],
    "Total Students" : [studentCount],
    "Total Budget" : [totalBudget],
    "Average Math Score" : [averageMathScore],
    "Average Reading Score" : [averageReadingScore],
    "Percent Passing Math" : [percentPassingMath],
    "Percent Passing Reading" : [percentPassingReading],
    "Percent Passing Math and Reading" : [percentPassingBoth],
    
})

#data formatting
districtSummary["Total Students"] = districtSummary["Total Students"].map("{:,}".format)
districtSummary["Total Budget"] = districtSummary["Total Budget"].map("{:,}".format)
districtSummary["Average Math Score"] = districtSummary["Average Math Score"].map("{:.2f}".format)
districtSummary["Average Reading Score"] = districtSummary["Average Reading Score"].map("{:.2f}".format)
districtSummary["Percent Passing Math"] = districtSummary["Percent Passing Math"].map("{:.2f}%".format)
districtSummary["Percent Passing Reading"] = districtSummary["Percent Passing Reading"].map("{:.2f}%".format)
districtSummary["Percent Passing Math and Reading"] = districtSummary["Percent Passing Math and Reading"].map("{:.2f}%".format)
districtSummary
```

# School Summary

```python
#school name
schoolDataMerged ["school_name"] .unique()
```

```python
schoolTypes = schoolData.set_index(["school_name"])["type"]
```

```python
#Total students Per SChool
studentsPerSchool = schoolDataMerged["school_name"].value_counts()
```

```python
#total budget per school 
budgetPerSchool = schoolDataMerged.groupby(["school_name"]).mean()["budget"]
```

```python
budgetPerStudent = budgetPerSchool/studentsPerSchool
```

```python
#average math scores per school
mathPerSchool =schoolDataMerged.groupby(["school_name"]).mean()["math_score"]
#average reading scores per school
readingPerSchool =schoolDataMerged.groupby(["school_name"]).mean()["reading_score"]
```

```python
#number of students passing math per school
PassMathPerSchool = schoolDataMerged[(schoolDataMerged["math_score"] >= 70)]
#number of students passing reading  per school
PassReadingPerSchool = schoolDataMerged[(schoolDataMerged["reading_score"] >= 70)]

# number of students passing math and reading per school
PassReadingAndMathPerSchool = schoolDataMerged[(schoolDataMerged["reading_score"] >= 70) &
                                       (schoolDataMerged["math_score"] >= 70)]
PassReadingAndMathPerSchool.head()
```

```python
#percent pasing math per school
schoolsPercentMath = (PassMathPerSchool.groupby(["school_name"]).count()["student_name"] / studentsPerSchool)*100
schoolsPercentMath
```

```python
#percent pasing reading per school
schoolsPercentReading = (PassReadingPerSchool.groupby(["school_name"]).count()["student_name"] / studentsPerSchool)*100
schoolsPercentReading
```

```python
#percent pasing math and reading per school
schoolsPercentMathReading = (PassReadingAndMathPerSchool.groupby(["school_name"]).count()["student_name"] / studentsPerSchool)*100
schoolsPercentMathReading 
```

```python
# Create data frame for school summery
summaryOfSchoolsData = pd.DataFrame({
    "School Type" : schoolTypes,
    "Total Students" : studentsPerSchool,
    "Total Budget" : budgetPerSchool,
    "Per Student Budget" : budgetPerStudent,
    "Average Math Score" : mathPerSchool,
    "Average Reading Score" : readingPerSchool,
   "% Passing Math" : schoolsPercentMath,
   "% Passing Reading" : schoolsPercentReading,
    "% Passing Math and Reading" : schoolsPercentMathReading
})

#data formatting
summaryOfSchoolsData["Total Students"] = summaryOfSchoolsData["Total Students"].map("{:,}".format)
summaryOfSchoolsData["Total Budget"] = summaryOfSchoolsData["Total Budget"].map("${:,.2f}".format)
summaryOfSchoolsData["Per Student Budget"] = summaryOfSchoolsData["Per Student Budget"].map("${:,.2f}".format)
summaryOfSchoolsData["Average Math Score"] = summaryOfSchoolsData["Average Math Score"].map("{:,.2f}".format)
summaryOfSchoolsData["Average Reading Score"] = summaryOfSchoolsData["Average Reading Score"].map("{:,.2f}".format)
summaryOfSchoolsData["% Passing Math"] = summaryOfSchoolsData["% Passing Math"].map("{:,.2f}%".format)
summaryOfSchoolsData["% Passing Reading"] = summaryOfSchoolsData["% Passing Reading"].map("{:,.2f}%".format)
summaryOfSchoolsData["% Passing Math and Reading"] = summaryOfSchoolsData["% Passing Math and Reading"].map("{:,.2f}%".format)        

summaryOfSchoolsData
```

# Highest Performing Schools

```python
HighPerformingSChools = summaryOfSchoolsData.sort_values(["% Passing Math and Reading"], ascending=False)
HighPerformingSChools.head()
```

# Lowest Performing Schools
